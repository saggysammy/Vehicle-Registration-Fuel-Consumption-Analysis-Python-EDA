```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,6)

df = pd.read_csv("C:\\Users\\hande\\Downloads\\vehicle_registrations_500k.csv")
df.head()
```

# Cleaning Steps

Removed duplicate records

Handled missing values

Ensured numeric consistency

```python

# Ensure numeric columns
df['vehicleCount'] = pd.to_numeric(df['vehicleCount'], errors='coerce')
df['grossVehicleWeight'] = pd.to_numeric(df['grossVehicleWeight'], errors='coerce')
df.describe()
```

```python
df.isnull()
```

# ðŸ” Exploratory Data Analysis (EDA)

```python
# Keep only valid years
df_clean = df[
    (df['registrationYear'] >= 2000) &
    (df['registrationYear'] <= 2025)
]

```

```python
yearly_reg = (
    df_clean
    .groupby('registrationYear')['vehicleCount']
    .sum()
    .sort_index()
)

plt.figure(figsize=(10, 5))

sns.lineplot(
    x=yearly_reg.index,
    y=yearly_reg.values,
    marker='o'
)

plt.title("Vehicle Registrations Trend Over Years")
plt.xlabel("Registration Year")
plt.ylabel("Total Vehicle Registrations")
plt.xticks(yearly_reg.index, rotation=45)
plt.tight_layout()
plt.show()
```

# ðŸ“Š Visual 2: Top 10 States by Vehicle Registrations

```python
top_states = df.groupby('stateName')['vehicleCount'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_states.values, y=top_states.index)
plt.title("Top 10 States by Vehicle Registrations")
plt.xlabel("Total Registrations")
plt.ylabel("State")
plt.show()
```

# ðŸ“Š Visual 3: Vehicle Registrations by Vehicle Category

```python
category_dist = df.groupby('vehicleCategoryName')['vehicleCount'].sum()

sns.barplot(x=category_dist.values, y=category_dist.index)
plt.title("Vehicle Registrations by Category")
plt.xlabel("Total Registrations")
plt.ylabel("Vehicle Category")
plt.show()
```

# ðŸ“Š Visual 4: Fuel Type Distribution (Pie Chart)

```python
# Aggregate fuel data
fuel_dist = (
    df.groupby('fuelName')['vehicleCount']
    .sum()
    .sort_values(ascending=False)
)

# Keep top 5 fuels, group rest as 'Others'
top_fuels = fuel_dist.head(5)
others = fuel_dist.iloc[5:].sum()

fuel_clean = pd.concat([top_fuels, pd.Series({'Others': others})])

# Plot clean pie chart
plt.figure(figsize=(7,7))
plt.pie(
    fuel_clean.values,
    labels=fuel_clean.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Fuel Type Distribution of Vehicles (Top Fuels)")
plt.show()

```

# ðŸ“Š Visual 5: Electric Vehicle Growth Over Years

```python
# Filter Electric Vehicles safely
ev_data = df[df['fuelName'].str.contains('electric', case=False, na=False)]

# Group by year
ev_trend = ev_data.groupby('registrationYear')['vehicleCount'].sum()

# Plot
sns.lineplot(
    x=ev_trend.index,
    y=ev_trend.values,
    marker='o'
)

plt.title("Electric Vehicle Registration Trend in India")
plt.xlabel("Registration Year")
plt.ylabel("EV Registrations")
plt.show()

```

# ðŸ“Š Visual 6: Monthly Registration Pattern

```python
monthly_reg = df.groupby('registrationMonthMMYY')['vehicleCount'].sum().head(12)

sns.barplot(x=monthly_reg.index, y=monthly_reg.values)
plt.title("Monthly Vehicle Registration Pattern")
plt.xlabel("Month")
plt.ylabel("Registrations")
plt.xticks(rotation=45)
plt.show()
```

# ðŸ“Š Visual 7: Sale Type vs Vehicle Registrations

```python
sale_type_dist = df.groupby('saleType')['vehicleCount'].sum()

sns.barplot(x=sale_type_dist.values, y=sale_type_dist.index)
plt.title("Vehicle Registrations by Sale Type")
plt.xlabel("Registrations")
plt.ylabel("Sale Type")
plt.show()


```

# ðŸ“Š Visual 8: Vehicle Category vs Fuel Type

```python
# Aggregate
cat_fuel = df.pivot_table(
    values='vehicleCount',
    index='vehicleCategoryName',
    columns='fuelName',
    aggfunc='sum',
    fill_value=0
)

# Keep only top categories and fuels to avoid clutter
top_categories = cat_fuel.sum(axis=1).sort_values(ascending=False).head(6).index
top_fuels = cat_fuel.sum(axis=0).sort_values(ascending=False).head(6).index

heatmap_data = cat_fuel.loc[top_categories, top_fuels]

plt.figure(figsize=(10,6))
sns.heatmap(
    heatmap_data,
    cmap='YlGnBu',
    annot=True,
    fmt='.0f'
)
plt.title("Vehicle Category vs Fuel Type (Top Categories & Fuels)")
plt.xlabel("Fuel Type")
plt.ylabel("Vehicle Category")
plt.tight_layout()
plt.show()

```

# ðŸ“Š Visual 9: Top 10 RTOs by Vehicle Registrations


```python
# Data
top_rto = (
    df.groupby('rtoName')['vehicleCount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=top_rto.values,
    y=top_rto.index,
    palette="Blues_r",      # shading
    edgecolor="black"       # clear bar edges
)

# Add value labels at end of bars
for i, value in enumerate(top_rto.values):
    ax.text(
        value,              # x-position
        i,                  # y-position
        f'{value:,}',       # formatted number
        va='center',
        ha='left',
        fontsize=10
    )

plt.title("Top 10 RTOs by Vehicle Registrations", fontsize=14)
plt.xlabel("Total Registrations")
plt.ylabel("RTO Name")
plt.tight_layout()
plt.show()

```

# ðŸ“Š Visual 10: Vehicle Class Distribution

```python
# Aggregate registrations by vehicle class
class_dist = (
    df.groupby('vehicleClassName')['vehicleCount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))

ax = sns.barplot(
    x=class_dist.values,
    y=class_dist.index,
    palette="Blues_r",     # shading
    edgecolor="black"      # bar borders
)

# Add numeric labels
for i, value in enumerate(class_dist.values):
    ax.text(
        value,
        i,
        f'{value:,}',
        va='center',
        ha='left',
        fontsize=10
    )

plt.title("Top 10 Vehicle Classes by Registrations", fontsize=14)
plt.xlabel("Total Registrations", fontsize=12)
plt.ylabel("Vehicle Class", fontsize=12)
plt.tight_layout()
plt.show()


```

# ðŸ“Š Visual 11: Pollution Norm vs Registrations

```python
# Aggregate registrations by pollution norm
pollution_dist = (
    df.groupby('pollutionNorm')['vehicleCount']
    .sum()
    .sort_values(ascending=False)
    .head(8)   # keep top norms only
)

plt.figure(figsize=(10, 6))

ax = sns.barplot(
    x=pollution_dist.values,
    y=pollution_dist.index,
    palette="Greens_r",    # green shade suits pollution context
    edgecolor="black"
)

# Add numeric labels
for i, value in enumerate(pollution_dist.values):
    ax.text(
        value,
        i,
        f'{value:,}',
        va='center',
        ha='left',
        fontsize=10
    )

plt.title("Top Pollution Norms by Vehicle Registrations", fontsize=14)
plt.xlabel("Total Registrations", fontsize=12)
plt.ylabel("Pollution Norm", fontsize=12)
plt.tight_layout()
plt.show()

```

# ðŸ“Š Visual 12: Correlation Heatmap (Numeric Features)
